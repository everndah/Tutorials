## What is Jupyter Notebook?

* It is a widely used web based application or a browser based notebook that is essential and is easy to use for data science projects since with it one can easily collaborate, share, and also publish their data science results.It is essential in  publishig results because with it one can combine code, images,comments ,plots etc.

> Refer to documentation https://jupyter-notebook.readthedocs.io/en/stable/ 

* It is browser based because when you start Jupyter notebook it will open in your default browser. 

* Jupyter Notebook supports over forty programming languages, some of the languages being Julia, Python, and R.

* Another thing to note is that it is possible to run Jupyter notebook on your computer without internet connection. But there are also notebooks that can run in the cloud such as Google’s Collaboratory project, and jupyter notebooks that run on Binder.

**If you would wish to install Jupyter notebook;**

* Jupyter notebooks can be used for all data science tasks and that is what  makes it one of the most useful tools in the data science community.That means that jupyter notebooks can be used for data cleaning, transformation,numerical simulation nad there is also exploratory data analysis, data visualization, statistical modeling, machine learning, deep learning among other tasks.


* While jupyter can run as a stand alone tool, Anaconda is one of the Data Science platforms or distributions that is prefferable when installing Jupyter notebooks.

## What is Anaconda?

* If you check Anaconda's website you’ll get to understand that it is a prepackaged data science software distribution which was built by data scientists, for data scientists and it comes with several packages.

> Packages are the tools that one uses to accomplish specific data science projects.  

* The tools that are found in Anaconda can be used for exploring datasets, modelling datasets and visualizing them. 
* Therefore to get started with your data science projects one of the suitable tools is anaconda. 
* Anaconda can be said to be an all in one store for data scientists.

* Some of the packages that come with it are Numpy, Pandas, Matplotlib, Sckit learn and Jupyter Notebooks.

## Why is Anaconda useful?

* If you need a tool box to do your data science projects then Anaconda is that tool box that has some of the popular packages.
